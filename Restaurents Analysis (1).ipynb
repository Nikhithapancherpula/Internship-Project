### **1. Plot the locations of restaurants on a map using longitude and latitude coordinates.**

```python
import folium

# Create a folium map centered at an arbitrary location

map_loc=[final_df['Latitude'].loc[0],final_df['Longitude'].iloc[0]]
my_map=folium.Map(location=map_loc,zoom_start=12)

# Add markers for each restaurant
for _, restaurant in final_df.iterrows():
    folium.Marker( location=[restaurant['Latitude'], restaurant['Longitude']],
                  popup=restaurant['Restaurant Name'] ).add_to(my_map)

# Save the map to an HTML file or display it
my_map.save("restaurant_map.html")
```

### **2. Identify any patterns or clusters of restaurants in specific areas.**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

features=final_df[['Longitude','Latitude', 'Price range','Aggregate rating', 'Votes']]

scaler=StandardScaler()
features_scaled = scaler.fit_transform(features)

#apply k-means Clustering
kmeans=KMeans (n_clusters = 3, random_state = 42)
final_df['Cluster']=kmeans.fit_predict(features_scaled)

#Use seaborn to visualization
sns.scatterplot(x='Longitude', y='Latitude', hue='Cluster', data=final_df, palette='viridis')
plt.title('Restaurant Clusters')
plt.show()
```

## **Votes Analysis**

### **1. Identify the restaurants with the highest and lowest number of votes**

```python
x=final_df[['Restaurant Name' , 'Votes']]
highest_votes=x[x['Votes']==x['Votes'].max()]
lowest_votes=x[x['Votes']==x['Votes'].min()]

print('Restaurants with highest votes:',highest_votes)
print('Restaurants with lowest_votes:',lowest_votes)
```
